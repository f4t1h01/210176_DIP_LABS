```python
import cv2
import matplotlib.pyplot as plt

# Read a grayscale image
image = cv2.imread('1.png', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and equalized images
plt.figure(figsize=(10,5))

# Original image and its histogram
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Function Explanation: ravel() flattens a multidimensional array into 1D.
# In this case, it flattens the 2D image matrix into a 1D array to plot its histogram.
plt.subplot(2, 2, 2)
plt.hist(image.ravel(), 256, [0,256])
plt.title('Histogram (Original)')

# Equalized image and its histogram
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')

plt.title('Equalized Image')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), 256, [0,256])
plt.title('Histogram (Equalized)')
plt.show()
```

Part 2: Histogram Equalization on Color Images

```python
import cv2
import matplotlib.pyplot as plt
# Read a color image
color_image = cv2.imread('1.png')
# Split the image into R, G, B channels
(b, g, r) = cv2.split(color_image)
# Equalize each channel separately
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)
# Merge the equalized channels
equalized_color_image = cv2.merge([equalized_b,
equalized_g, equalized_r])
# Display the original and equalized images
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(color_image,
cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_color_image,
cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image (Channels Separately)')
plt.axis('off')
plt.show()
```

Part 3: Applying Histogram matching

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt
# Read two grayscale images
image_source = cv2.imread('1.png',cv2.IMREAD_GRAYSCALE)
image_target = cv2.imread('bright.png',cv2.IMREAD_GRAYSCALE)
# Function Explanation: exposure.match_histograms() matches the histogram of the source image
# to that of the target image. It adjusts pixel values to make the intensity distribution
# of the source image resemble the target image's histogram.
# Perform histogram matching
matched_image =exposure.match_histograms(image_source,image_target)
# Display the source, target, and matched images
plt.figure(figsize=(15,5))
```

```python
# Source image and its histogram
plt.subplot(3, 3, 1)
plt.imshow(image_source, cmap='gray')
plt.title('Source Image')
plt.axis('off')
plt.subplot(3, 3, 2)
plt.hist(image_source.ravel(), 256, [0,256])
plt.title('Histogram (Source)')
# Target image and its histogram
plt.subplot(3, 3, 4)
plt.imshow(image_target, cmap='gray')
plt.title('Target Image')
plt.axis('off')
plt.subplot(3, 3, 5)
plt.hist(image_target.ravel(), 256, [0,256])
plt.title('Histogram (Target)')
# Matched image and its histogram
plt.subplot(3, 3, 7)
plt.imshow(matched_image, cmap='gray')
plt.title('Matched Image')
plt.axis('off')
plt.subplot(3, 3, 8)
plt.hist(matched_image.ravel(), 256, [0,256])
plt.title('Histogram (Matched)')
plt.show()
```

```python
# =================================================================
# Lab 5: Intensity Transformation II (Histogram Processing)
# Name: [Your Name Here]
# Student ID: [Your ID Here]
# Date: [Current Date]
# =================================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure

# --- PREPARATION ---
# Note: Ensure you have uploaded your images to the Colab 'Files' tab first.
# Replace 'image1.jpg' and 'image2.jpg' with your actual filenames.
source_img_path = '1.png' # Corrected from '1.jpg'
target_img_path = 'bright.png' # Corrected from 'bright.jpg'

# Load images
img_bgr = cv2.imread(source_img_path)
target_bgr = cv2.imread(target_img_path)

# Add checks for successful image loading
if img_bgr is None:
    print(f"Error: Could not load source image from {source_img_path}. Please check the file path and ensure the image exists.")
    raise FileNotFoundError(f"Source image not found: {source_img_path}")
if target_bgr is None:
    print(f"Error: Could not load target image from {target_img_path}. Please check the file path and ensure the image exists.")
    raise FileNotFoundError(f"Target image not found: {target_img_path}")

# Convert to RGB for proper display in Matplotlib
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
target_rgb = cv2.cvtColor(target_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

def plot_img_and_hist(image, title):
    """Helper function to plot image and its flattened histogram."""
    plt.figure(figsize=(12, 4))

    # Plot Image
    plt.subplot(1, 2, 1)
    if len(image.shape) == 2:
        plt.imshow(image, cmap='gray')
    else:
        plt.imshow(image)
    plt.title(f"{title} Image")
    plt.axis('off')

    # Plot Histogram using ravel() as required by Lab manual
    plt.subplot(1, 2, 2)
    plt.hist(image.ravel(), bins=256, range=(0, 256), color='black', alpha=0.7)
    plt.title(f"{title} Histogram")
    plt.xlabel('Intensity Value')
    plt.ylabel('Frequency')
    plt.show()

# -----------------------------------------------------------------
# PART 1: Histogram Equalization (Grayscale)
# -----------------------------------------------------------------
print("--- PART 1: Grayscale Histogram Equalization ---")
equalized_gray = cv2.equalizeHist(img_gray)
plot_img_and_hist(img_gray, "Original Grayscale")
plot_img_and_hist(equalized_gray, "Equalized Grayscale")

# -----------------------------------------------------------------
# PART 2: Histogram Equalization (Color Images)
# -----------------------------------------------------------------
print("\n--- PART 2: Color Histogram Equalization ---")
# To equalize color, convert to YCrCb and equalize only the 'Y' (Luminance) channel
ycrcb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2YCrCb)
ycrcb[:, :, 0] = cv2.equalizeHist(ycrcb[:, :, 0])
equalized_color_rgb = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2RGB)

plot_img_and_hist(img_rgb, "Original Color")
plot_img_and_hist(equalized_color_rgb, "Equalized Color")

# -----------------------------------------------------------------
# PART 3: Histogram Matching
# -----------------------------------------------------------------
print("\n--- PART 3: Histogram Matching ---")
# Match the source image's histogram to the target image's histogram
matched_img = exposure.match_histograms(img_rgb, target_rgb, channel_axis=-1)

# Display Source, Target, and Matched result
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(img_rgb); plt.title('Source Image'); plt.axis('off')
plt.subplot(1, 3, 2); plt.imshow(target_rgb); plt.title('Target Image'); plt.axis('off')
plt.subplot(1, 3, 3); plt.imshow(matched_img); plt.title('Matched Result'); plt.axis('off')
plt.show()

plot_img_and_hist(matched_img, "Matched Result")

# -----------------------------------------------------------------
# PART 4: Combining Histogram Equalization and Matching
# ----------------
```
